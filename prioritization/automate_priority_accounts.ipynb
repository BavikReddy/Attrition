```python
import pandas as pd
import numpy as np
```

```python
#load final_parent_priority
final_parent_priority_df = pd.read_csv('data/prioritization_data/final_parent_priority.csv')
```

```python
final_parent_priority_df.columns
```

```python
final_parent_priority_df.shape
```

```python
len(final_parent_priority_df['acctcode'].unique())
```

```python
account_level_df = final_parent_priority_df[['parentname_x', 'acctcode', 'attrition_prob_x',
       'last_transaction_week', 'last_transaction_month',
       'count_sf_cases_last3m', 'phone', 'mobilephone', 'otherphone',
       'Current_account_managed', 'last_52_weeks_gallons_y',
       'monthly_avg_gallons_last_52_weeks_y', 'max_gallons_per_month_y',
       'last_52_weeks_gallons_max_y', 'last_52_weeks_revenue_y',
       'monthly_avg_revenue_last_52_weeks_y', 'max_revenue_per_month_y',
       'last_52_weeks_revenue_max_y', 'last_4_weeks_gallons','last_8_weeks_gallons',
        'last_13_weeks_gallons','last_26_weeks_gallons']]
```

```python
#Identify all unique parents which has more than 1 accounts
duplicated_parents_list = list(account_level_df[account_level_df['parentname_x'].duplicated()]['parentname_x'].unique())
```

```python
len(duplicated_parents_list)
```

```python
#Access all accounts info from duplicated_parents_list
duplicated_accounts_df = account_level_df[account_level_df['parentname_x'].isin(duplicated_parents_list)]
```

```python
duplicated_accounts_df
```

```python
combined_accounts_df = duplicated_accounts_df.copy()
```

#### Create new columns which are required to prioritize

```python
combined_accounts_df['decline%_26-4'] = ((combined_accounts_df['last_26_weeks_gallons'] - combined_accounts_df['last_4_weeks_gallons'])/ combined_accounts_df['last_26_weeks_gallons'])*100
```

```python
combined_accounts_df['valid_phone'] = np.where(combined_accounts_df['phone'].notna(),1,2)
```

```python
combined_accounts_df['rank_last_transaction_week'] = combined_accounts_df.groupby('parentname_x')['last_transaction_week'].rank(method='dense', ascending=False)
```

```python
combined_accounts_df['rank_revenue'] = combined_accounts_df.groupby('parentname_x')['last_52_weeks_revenue_y'].rank(method='dense', ascending=False)
```

```python
combined_accounts_df['rank_gallons'] = combined_accounts_df.groupby('parentname_x')['last_52_weeks_gallons_y'].rank(method='dense', ascending=False)
```

```python
combined_accounts_df['rank_sf_cases'] = combined_accounts_df.groupby('parentname_x')['count_sf_cases_last3m'].rank(method='dense', ascending=False)
```

```python
combined_accounts_df['rank_decline%'] = combined_accounts_df.groupby('parentname_x')['decline%_26-4'].rank(method='dense', ascending=False)
```

```python
combined_accounts_df[['valid_phone','rank_last_transaction_week','rank_decline%','rank_sf_cases']]
```

#### Calculate Final Rank

```python
# Calculate global_rank by taking the average of rank columns
combined_accounts_df['global_rank_mean'] = combined_accounts_df[['valid_phone','rank_last_transaction_week','rank_decline%','rank_sf_cases']].fillna(0).sum(axis=1) / 4
```

```python
combined_accounts_df['global_rank_mean']
```

```python
# Calculate Final_rank column by grouping parentnames and prioritize 
combined_accounts_df['account_rank'] = combined_accounts_df.groupby('parentname_x')['global_rank_mean'].rank(method='dense', ascending=True)
```

```python
combined_accounts_df['account_rank']
```

```python
combined_accounts_df.shape
```

```python
combined_accounts_df.columns
```

```python
account_priority_df = combined_accounts_df[['acctcode','account_rank']]
```

#### Merge account priority df with original parent priority df

```python
parent_account_priority_df = pd.merge(final_parent_priority_df,account_priority_df, on ='acctcode',how='left')
```

```python
#fill missing account_rank with 1
parent_account_priority_df['account_rank'] = parent_account_priority_df['account_rank'].fillna(1)
```

```python
#drop columns which are not required
parent_account_priority_df.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0','parentname_y', 'attrition_prob_y'],inplace = True)
```

```python
parent_account_priority_df
```

```python
parent_account_priority_df.columns
```

```python
#Rename columns
# Replace '_x' with '_parent_level' and '_y' with '_account_level' in column names
updated_columns = [
    col.replace('_x', '_parent_level').replace('_y', '_account_level') 
    if col not in ['parentname_x', 'attrition_prob_x'] 
    else col.replace('_x', '')
    for col in parent_account_priority_df.columns
]
```

```python
updated_columns
```

```python
parent_account_priority_df.columns = updated_columns
```

```python
parent_account_priority_df
```
